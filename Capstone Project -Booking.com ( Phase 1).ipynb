### Creating Data1.csv 

```python
london_url=[]
start = 0
end = 10
page_url = url[0]
driver.get(page_url)
full=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[4]/div[2]/div/div[2]/div[4]/div[11]/a/span')
full[0].click()
time.sleep(10)
```

```python
for i in range(start, end):
    for i in driver.find_elements(By.XPATH,'//h3[@class="aab71f8e4e"]/a'):
        london_url.append(i.get_attribute('href'))
```

```python
london_url
```

```python
len(london_url)
```

```python
hotel_name = []
cities = []
location = []
no_of_reviews = []
user_rating = []

for i in london_url:
    driver.get(i)
    time.sleep(10)
    
    h=driver.find_elements(By.XPATH,'//h2[@class="d2fee87262 pp-header__title"]')
    for element in h:
        try:
            hotel_name.append(element.text)
        except NoSuchElementException:
            hotel_name.append("-")
        
    c=driver.find_elements(By.XPATH,'//li[@class="bui-breadcrumb__item hp-breadcrumb__item"][4]/div')
    for element in c:
        try:
            cities.append(element.text)
        except NoSuchElementException:
            cities.append("-")
    
    
    l=driver.find_elements(By.XPATH,'//p[@class="address address_clean"]/span[1]')
    for element in l:
        try:
            location.append(element.text)
        except NoSuchElementException:
            location.append("-")
    
    n=driver.find_elements(By.XPATH,'//span[@class="a3b8729ab1 f45d8e4c32 d935416c47"]')
    for element in n:
        try:
            no_of_reviews.append(element.text)
        except NoSuchElementException:
            no_of_reviews.append("-")
    
    u=driver.find_elements(By.XPATH,'//div[@class="b2b990caf1"]/button/div/div/div[1]')
    for element in u:
        try:
            user_rating.append(element.text)
        except NoSuchElementException:
            user_rating.append("-")
```

```python
for i in tokyo_url:
    driver.get(i)
    time.sleep(10)
    
    h = driver.find_elements(By.XPATH, '//h2[@class="d2fee87262 pp-header__title"]')
    for element in h:
        try:
            hotel_name.append(element.text)
        except NoSuchElementException:
            hotel_name.append("-")

    try:
        r = driver.find_elements(By.XPATH, '/html/body/div[3]/div/div[5]/div[1]/div[2]/div[2]/div/div/div/div/div/ul/li[2]/div[2]/div')
        if len(r) > 0:
            for element in r:
                try:
                    room_type.append(element.text)
                except NoSuchElementException:
                    room_type.append("-")
        else:
            room_type.append("-")
    except Exception as r:
        print("An exception occurred:", str(r))
    
    
    try:
        sb=driver.find_elements(By.XPATH,'/html/body/div[3]/div/div[5]/div[1]/div[2]/div[3]/div/div/div[1]/div[2]/div/div/div/ul/li[2]/div/div/div/span/div/span')
        if len(sb) > 0:
            for element in sb:
                try:
                    single_bed.append("Yes")
                except NoSuchElementException:
                    single_bed.append("No")
        else:
            single_bed.append("No")
    except Exception as sb:
        print("An exception occurred:", str(sb))

    try:
        db = driver.find_elements(By.XPATH, '/html/body/div[3]/div/div[5]/div[1]/div[2]/div[3]/div/div/div[1]/div[2]/div/div/div/ul/li[2]/div/div/div/span/div/span')
        if len(db) > 0:
            for element in db:
                try:
                    double_bed.append("Yes")
                except NoSuchElementException:
                    double_bed.append("No")
        else:
            double_bed.append("No")
    except Exception as db:
        print("An exception occurred:", str(db))
    
    
    try:
        p = driver.find_elements(By.XPATH, '/html/body/div[3]/div/div[5]/div[1]/div[2]/div[11]/div/div[8]/div/div/section/div/div[2]/div[2]/div[2]/ul/li[1]/div/span/div/div[1]')
        if len(p) > 0:
            for element in p:
                try:
                    price.append(element.text)
                except NoSuchElementException:
                    price.append("-")
        else:
            price.append("-")
    except Exception as p:
        print("An exception occurred:", str(p))
```
