```python
# Write a query to count the occurrences of each breakfast cuisine type.
```

```python
query_count_cuisine = """
SELECT breakfast_cuisine, COUNT(*) AS cuisine_count
FROM (
    SELECT 
        CASE 
            WHEN breakfast_cusine1 != '-' THEN breakfast_cusine1
            WHEN breakfast_cusine2 != '-' THEN breakfast_cusine2
            WHEN breakfast_cusine3 != '-' THEN breakfast_cusine3
            ELSE '-'
        END AS breakfast_cuisine
    FROM DATA2csv
)
WHERE breakfast_cuisine != '-'
GROUP BY breakfast_cuisine;
"""

# Execute the SQL query
cur.execute(query_count_cuisine)
for i in result:
    print(i)
```

```python
# Write a query to retrieve hotels that offer a certain(International )  breakfast cuisine
```

```python
result=cur.execute("SELECT hotel_name FROM DATA2csv WHERE breakfast_cusine1 = 'International' OR breakfast_cusine2 = 'International' OR breakfast_cusine3 = 'International'")
for i in result:
    print(i)
```

```python
# Write a query to see whether an hotel has all amenities such as Heating, Lift and Bar
```

## From here we will be making SUBQUERIES including the datafrom table 1/Table 2/Table 3 

```python
result = cur.execute("""
    SELECT DISTINCT DATA1csv.hotel_name, DATA1csv.user_rating, DATA2csv.free_wifi 
    FROM DATA1csv 
    INNER JOIN DATA2csv ON DATA1csv.hotel_name = DATA2csv.hotel_name 
    WHERE CAST(DATA1csv.user_rating AS FLOAT) > 5 AND DATA2csv.free_wifi != '-'
""")

for i in result:
    print(i)
```

```python
# Write a query to find hotels that offer both family rooms and non-smoking rooms(Use table 1 and Table 2)
```

```python
result=cur.execute("""SELECT DISTINCT DATA1csv.hotel_name
FROM DATA1csv DATA1csv
JOIN DATA2csv DATA2csv ON DATA1csv.hotel_name = DATA1csv.hotel_name
WHERE DATA2csv.family_room = 'Yes' AND DATA2csv.non_smoking_room = 'Yes';
""")
for i in result:
    print(i)
```

```python
# Write a query to calculate the average price for each room type (Use table2 and Table 3)
```

```python
result=cur.execute("""SELECT DISTINCT DATA3csv.room_type, AVG(DATA3csv.price) AS average_price
FROM DATA2csv DATA2csv
JOIN DATA3csv DATA3csv ON DATA2csv.hotel_name = DATA3csv.hotel_name
GROUP BY DATA3csv.room_type
""")
for i in result:
    print(i)
```
